```python
def load_benchmark():
    try:
        with open('Book1and2_pair.csv', 'r') as f:
            benchmark = f.read().splitlines()
        if benchmark is None:
            print("no file.")
        return benchmark
    except:
        print("Error occurred. Check if file exists")
        
def calc_measure(benchmark, results):
    if len(results) == 0:
        print('Precision = 0, Recall = 0, Fmeasure = 0')
        return
    count = 0
    for pair in results:
        if pair in benchmark:
            count = count + 1
    if count == 0:
        print('Precision=0, Recall=0, Fmeasure=0')
        return
    precision = count / len(results)
    recall = count / len(benchmark)
    f_measure = 2 * precision * recall / (precision + recall)
    print("Count=", count, ", len(results)=", len(results), ", len(benchmark)=", len(benchmark))
    print("Precision=", precision, ", Recall=", recall, ", Fmeasure=", f_measure)
    return
```

```python
def nested_loop_by_name_jaccard():
    results = []
    id1 = 0
    id2 = 0
    book_title1 = None
    book_title2 = None    # Nguyen Van Quoc Chuong 47787810
    threshold = 0.75
    q = 3
    start_time = datetime.datetime.now()
    for i in range(0, book1.shape[0]): # the number of book1 rows
        id1 = book1.id[i]
        book_title1 = book1.title[i]
        
        for j in range(0, book2.shape[0]): # the number of book2 rows
            id2 = book2.id[j]
            book_title2 = book2.book_title[j]
            sim = calc_jaccard(str(book_title1), str(book_title2), q)
            #print("sim:", sim)
            if sim >= threshold:
                results.append(str(id1) + ',' + str(id2))
    end_time = datetime.datetime.now()
    time = end_time - start_time
    print("q = ", q, "          threshold = ", threshold)
    print("Total Time:", round(time.total_seconds() * 1000, 3), 'milliseconds')
    return results
```
